```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_track=pd.read_csv("C:\\Users\\Admin\\OneDrive\\Desktop\\Practise\\Spotify\\tracks1.csv")
df_track.head()
```

```python
df_track['release_date'] = df_track['release_date'].astype('datetime64[ns]')
df_track
```

```python
df_track=df_track.set_index('release_date')
df_track
```

```python
df_track['duration']=df_track['duration_ms'].apply(lambda x: round(x/1000))
df_track.drop("duration_ms" , inplace=True,axis=1 )
```

```python
df_track.head()
```

```python
corr_df1 = df_track.drop(['key','mode','explicit'] , axis =1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df1,annot=True,fmt='.1g',vmin=-1,vmax=1,center=0,cmap="mako",linewidths=1,linecolor="Red")
heatmap.set_title("Correlations heat map ")
heatmap.set_xticklabels(heatmap.getxticklabels(),rotation=90)
```

```python
sample_df=df_track.sample(int(0.004*len(df_track)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='loudness',x='energy',color='c').set(title="loudness vs energy")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='popularity',x='acousticness',color='r').set(title="popularity vs acousticness")
```

```python
df_track['dates']=df_track.index.get_level_values('release_date')
df_track.dates=pd.to_datetime(df_track.dates)
years=df_track.dates.dt.year
```

```python
total_dr=df_track.duration
fig_dims = (18,7)
fig ,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_dr,ax=ax,errwidth=False).set(title='Year vs Duration')
plt.xticks(rotation=90)
```
